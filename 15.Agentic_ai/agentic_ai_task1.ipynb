```python
import os
import requests
from langchain_google_genai import ChatGoogleGenerativeAI
# from langchain_groq import ChatGroq
from dotenv import load_dotenv

load_dotenv()
```

```python
print(f"API Key found: {'Yes' if os.getenv('GOOGLE_API_KEY') else 'No'}")
```

```python
from langchain_community.tools import DuckDuckGoSearchRun

search_tool = DuckDuckGoSearchRun()

result = search_tool.invoke("top new in india today")
```

```python
result
```

```python
api=os.getenv('GOOGLE_API_KEY').split(' ')[0]
```

```python
# llm = ChatGroq(model="llama-3.1-8b-instant")
llm = ChatGoogleGenerativeAI(model="gemini-2.5-flash", api_key=api)
```

```python
llm.invoke("hi")
```

```python
from langchain.agents import create_react_agent, AgentExecutor
from langchain import hub
```

```python
prompt = hub.pull("hwchase17/react")
```

```python
agent = create_react_agent(
    llm=llm,
    tools=[search_tool],
    prompt=prompt
)
```

```python
agent_executor = AgentExecutor(
    agent=agent,
    tools=[search_tool],
    verbose=True,
)
```

```python
response = agent_executor.invoke({"input": "3 ways to reach goa from delhi"})
print(response)
```
